# Введение

Цель работы — подобрать подходящее модельное распределение для переменной
характеристики генеральной совокупности.

Подключаем библиотеки

```r
install.packages("dplyr")
library(dplyr)
install.packages("fitdistrplus")
library(fitdistrplus)
install.packages("dgof")
library(dgof)
```

Берём данные из файла

```r
ds0 <- read.csv("AD.ds2.csv")
head(ds0)
```

```r
ds <- dplyr::select(ds0, s15 = v15)
ds
```

```r
fdBeta <- fitdist(ds$s15, "beta")
fdBeta
```

# Проверка с помощью теста

Проверяем с помощью теста Колмогорова-Смирнова подходит ли данное распределение для описания исследуемой совокупности.

Вычислим переменную, содержащую параметры бета-распределения и выведим ее значения

```r
fdBeta
```

С помощью функции ks.test выполним тест Колмогорова-Смирнова

```r
beta_p <- fdBeta$estimate
ks.test(ds$s15, "pbeta", beta_p[1], beta_p[2])
```

Наблюдаемое значение меньше критического(p-value больше 0.05), гипотиза о том, что проверяемое распределение исследуемой переменной характеристики генеральной совокупности не отвергается.

Вычислим переменную, содержащую параметры экспоненциального распределения и выведим ее значения

```r
fdExp
```

С помощью функции ks.test выполним тест Колмогорова-Смирнова

```r
exp_p <- fdExp$estimate
ks.test(ds$s15, "pexp", exp_p[1])
```

Наблюдаемое значение меньше критического(p-value больше 0.05), гипотиза о том, что проверяемое распределение исследуемой переменной характеристики генеральной совокупности не отвергается.

Вычислим переменную, содержащую параметры гамма-распределения и выведим ее значения

```r
fdGamma
```

С помощью функции ks.test выполним тест Колмогорова-Смирнова

```r
gamma_p <- fdGamma$estimate
ks.test(ds$s15, "pgamma", gamma_p[1], gamma_p[2])
```

Наблюдаемое значение меньше критического(p-value больше 0.05), гипотиза о том, что проверяемое распределение исследуемой переменной характеристики генеральной совокупности не отвергается.

Вычислим переменную, содержащую параметры распределения Вейбулла и выведим ее значения

```r
fdWeibull
```

С помощью функции ks.test выполним тест Колмогорова-Смирнова

```r
weibull_p <- fdWeibull$estimate
ks.test(ds$s15, "pweibull", weibull_p[1], weibull_p[2])
```

Наблюдаемое значение меньше критического(p-value больше 0.05), гипотиза о том, что проверяемое распределение исследуемой переменной характеристики генеральной совокупности не отвергается.

**Вывод:** бета-распределение с параметрами shape1 = 0.9255814 и shape2 = 8.2884232 подходит для описания исследуемой генеральной совокупности;

экспоненциальное распределение с параметром rate = 9.974481 подходит для описания исследуемой генеральной совокупности;

гамма-распределение с параметрами shape = 1.007521 и rate = 10.049350 подходит для описания исследуемой генеральной совокупности;
